```python
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df = pd.read_csv(url)
```

```python
print("Dimensioni del dataset:", df.shape)
```

```python
print("Informazioni sul dataset:")
print(df.info())
```

```python
df_cleaned = df.dropna(subset=['continent'])
```

```python
continent_cases = df_cleaned.groupby('continent')['total_cases'].max()
```

```python
print("Numero di casi totali per ogni continente:")
print(continent_cases)
```

```python
def calculate_statistics(continent): 
    total_world_cases = df['total_cases'].max()
    continent_cases = df[df['continent'] == continent]['total_cases']

    max_cases_continent = continent_cases.max() if not continent_cases.empty else 0
    avg_cases_continent = continent_cases.mean() if not continent_cases.empty else 0

    percent_continent = (continent_cases.sum() / total_world_cases) * 100
    return {
        'max_cases': max_cases_continent,
        'avg_cases': avg_cases_continent,
        'percent_total': percent_continent
    }
```

```python
def confront_statistics(continent1, continent2):
    stats_continent1 = calculate_statistics(continent1)
    stats_continent2 = calculate_statistics(continent2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df = pd.read_csv(url)
```

```python
italy_data_2022 = df[(df['location'] == 'Italy') & (df['date'].str.startswith('2022'))]
```

```python
italy_data_2022['date'] = pd.to_datetime(italy_data_2022['date'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(italy_data_2022['date'], italy_data_2022['total_cases'], marker='o', linestyle='-')
plt.title('Evoluzione dei casi totali in Italia nel 2022')
plt.xlabel('Data')
plt.ylabel('Casi totali')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df = pd.read_csv(url)
```

```python
selected_countries = ['Italy', 'Germany', 'France']
start_date = '2022-05-01'
end_date = '2023-04-30'
filtered_data = df[(df['location'].isin(selected_countries)) & (df['date'] >= start_date) & (df['date'] <= end_date)]
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([filtered_data[filtered_data['location'] == country]['icu_patients'] for country in selected_countries], labels=selected_countries)
plt.title('Differenza nel numero di pazienti in terapia intensiva (ICU) - Maggio 2022 / Aprile 2023')
plt.xlabel('Paese')
plt.ylabel('Numero di pazienti ICU')
plt.grid(True)
plt.tight_layout()
plt.show()
```
